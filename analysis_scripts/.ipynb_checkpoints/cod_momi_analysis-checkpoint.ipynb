```python
#Import momi2
import momi
```

```python
#let's visualize a model first - just a 2-pop split
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=5e4)
model.add_leaf("Norway", N=5e4)
model.move_lineages("Canada", "Norway", t=2.5e4, N=5e4)
```

```python
%matplotlib inline

yticks = [2.5e4, 5e4, 10e4]

fig = momi.DemographyPlot(
    model, ["Canada", "Norway"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#now with ancient migration
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=5e4)
model.add_leaf("Norway", N=5e4)
model.move_lineages("Canada", "Norway", t=2e4, p=.03)
model.move_lineages("Canada", "Norway", t=2.5e4, N=5e4)
```

```python
%matplotlib inline

yticks = [1e4, 2.5e4, 5e4, 7.5e4, 1e5]

fig = momi.DemographyPlot(
    model, ["Canada", "Norway"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#now with ancient + recent migration
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=5e4)
model.add_leaf("Norway", N=5e4)
model.move_lineages("Canada", "Norway", t=2e4, p=.03)
model.move_lineages("Canada", "Norway", t=5e1, p=.01)
model.move_lineages("Canada", "Norway", t=2.5e4, N=5e4)
```

```python
%matplotlib inline

yticks = [1e4, 2.5e4, 5e4, 7.5e4, 1e5]

fig = momi.DemographyPlot(
    model, ["Canada", "Norway"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#now with ancient + recent migration + recent pop size decline
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=5e5,g=-1e-4)
model.add_leaf("Norway", N=5e5,g=-1e-4)
model.set_size("Canada", g=0, t=1e2)
model.set_size("Norway", g=0, t=1e2)
model.move_lineages("Canada", "Norway", t=2e4, p=.03)
model.move_lineages("Canada", "Norway", t=5e1, p=.01)
model.move_lineages("Canada", "Norway", t=2.5e4, N=5e4)
```

```python
%matplotlib inline

yticks = [100, 200]

fig = momi.DemographyPlot(
    model, ["Canada", "Norway"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#with recent pop size decline - something like this
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=1e5,g=-5e-2)
model.add_leaf("Norway", N=1e5,g=-5e-2)
model.set_size("Canada", g=0, t=1e2)
model.set_size("Norway", g=0, t=1e2)
model.move_lineages("Canada", "Norway", t=8e2, p=.03)
model.move_lineages("Canada", "Norway", t=1e2, p=.01)
model.move_lineages("Canada", "Norway", t=1e3, N=5e5)

%matplotlib inline

yticks = [100, 200]

fig = momi.DemographyPlot(
    model, ["Canada", "Norway"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#add an intermediate "ghost" population and continuous migration (multiple pulses) through ghost
#vicariance at 128 kya (Carr + Marshall 2008)
#for inference we would set vicariance times + pulses, possibly population sizes, and then infer pulse strength? 
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=5e4)
model.add_leaf("Norway", N=5e4)
model.add_leaf("Ghost", N=5e4)
model.move_lineages("Canada", "Ghost", t=1e5, p=.01)
model.move_lineages("Canada", "Ghost", t=7.5e4, p=.01)
model.move_lineages("Canada", "Ghost", t=5e4, p=.01)
model.move_lineages("Canada", "Ghost", t=2.5e4, p=.01)
model.move_lineages("Ghost", "Canada", t=1e5, p=.01)
model.move_lineages("Ghost", "Canada", t=7.5e4, p=.01)
model.move_lineages("Ghost", "Canada", t=5e4, p=.01)
model.move_lineages("Ghost", "Canada", t=2.5e4, p=.01)
model.move_lineages("Norway", "Ghost", t=1e5, p=.01)
model.move_lineages("Norway", "Ghost", t=7.5e4, p=.01)
model.move_lineages("Norway", "Ghost", t=5e4, p=.01)
model.move_lineages("Norway", "Ghost", t=2.5e4, p=.01)
model.move_lineages("Ghost", "Norway", t=1e5, p=.01)
model.move_lineages("Ghost", "Norway", t=7.5e4, p=.01)
model.move_lineages("Ghost", "Norway", t=5e4, p=.01)
model.move_lineages("Ghost", "Norway", t=2.5e4, p=.01)
model.move_lineages("Canada", "Ghost", t=1.14e5, N=5e4)
model.move_lineages("Ghost", "Norway", t=1.28e5, N=5e4)


%matplotlib inline

yticks = [5e4, 1e5, 1.5e5, 2e5, 2.5e5]

fig = momi.DemographyPlot(
    model, ["Norway", "Ghost", "Canada"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#continuous migration (multiple pulses) without ghost
#vicariance at 128 kya (Carr + Marshall 2008)
#for inference we would set vicariance times + pulses, possibly population sizes, and then infer pulse strength? 
model = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
model.add_leaf("Canada", N=5e4)
model.add_leaf("Norway", N=5e4)
model.move_lineages("Canada", "Norway", t=1e5, p=.01)
model.move_lineages("Canada", "Norway", t=7.5e4, p=.01)
model.move_lineages("Canada", "Norway", t=5e4, p=.01)
model.move_lineages("Canada", "Norway", t=2.5e4, p=.01)
model.move_lineages("Norway", "Canada", t=1e5, p=.01)
model.move_lineages("Norway", "Canada", t=7.5e4, p=.01)
model.move_lineages("Norway", "Canada", t=5e4, p=.01)
model.move_lineages("Norway", "Canada", t=2.5e4, p=.01)
model.move_lineages("Canada", "Norway", t=1.28e5, N=5e4)

%matplotlib inline

yticks = [5e4, 1e5, 1.5e5, 2e5, 2.5e5]

fig = momi.DemographyPlot(
    model, ["Norway", "Canada"],
    figsize=(6,8),
    major_yticks=yticks)
```

```python
#initialize model with continuous migration (4 pulses), fitting the most recent datasets
CanNor4pulsemodel = momi.DemographicModel(N_e=5e4, gen_time=7,
                              muts_per_gen=1.25e-8)
```

```python
#set the data
CanNor4pulsemodel.set_data(sfs)
```

```python
#set the parameters - all random initial values
CanNor4pulsemodel.add_size_param("n_can",lower=1e3,upper=2e5)
CanNor4pulsemodel.add_size_param("n_nor",lower=1e3,upper=2e5)
CanNor4pulsemodel.add_pulse_param("p_pulse", upper=.4)
```

```python
#add demographic events
CanNor4pulsemodel.add_leaf("Can13",t=1,N="n_can")
CanNor4pulsemodel.add_leaf("Lof14",t=0,N="n_nor")
CanNor4pulsemodel.move_lineages("Can13", "Lof14", t=1e5, p="p_pulse")
CanNor4pulsemodel.move_lineages("Can13", "Lof14", t=7.5e4, p="p_pulse")
CanNor4pulsemodel.move_lineages("Can13", "Lof14", t=5e4, p="p_pulse")
CanNor4pulsemodel.move_lineages("Can13", "Lof14", t=2.5e4, p="p_pulse")
CanNor4pulsemodel.move_lineages("Lof14", "Can13", t=1e5, p="p_pulse")
CanNor4pulsemodel.move_lineages("Lof14", "Can13", t=7.5e4, p="p_pulse")
CanNor4pulsemodel.move_lineages("Lof14", "Can13", t=5e4, p="p_pulse")
CanNor4pulsemodel.move_lineages("Lof14", "Can13", t=2.5e4, p="p_pulse")
CanNor4pulsemodel.move_lineages("Can13", "Lof14", t=1.28e5, N="n_can")
```

```python
CanNor4pulsemodel.optimize(method="TNC")
###KERNELDEATH!!!
##try a simpler model? or run on hpc?
```
